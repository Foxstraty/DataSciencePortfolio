```python
# !pip3 install opencv-python

import cv2
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score, f1_score, recall_score
import cv2
import os

#getting current directory and defining the folder paths
cwd = os.getcwd()
folder_path1 = f'{cwd}/train/real'
folder_path2 = f'{cwd}/train/fake'
folder_path3 = f'{cwd}/test/real'
folder_path4 = f'{cwd}/test/fake'

# empty arrays
train_live = []
train_fake = []

count = 0
# Loop through all the files in the folder
for filename in os.listdir(folder_path1):
    # Check if the file is an image
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image file using OpenCV
        img = cv2.imread(os.path.join(folder_path1, filename))
        img = cv2.resize(img, (64, 64))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img.flatten()
        train_live.append(img)
        count += 1
        if count == 150:
            break
count = 0
for filename in os.listdir(folder_path2):
    # Check if the file is an image
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image file using OpenCV
        img = cv2.imread(os.path.join(folder_path2, filename))
        img = cv2.resize(img, (64, 64))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img.flatten()
        train_fake.append(img)
        count += 1
        if count == 150:
            break

X_train = np.concatenate((train_live, train_fake))
y_train = np.concatenate((np.ones(len(train_live)), np.zeros(len(train_fake))))

# Load the testing dataset
test_live = []
test_fake = []

count = 0
# Loop through all the files in the folder
for filename in os.listdir(folder_path3):
    # Check if the file is an image
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image file using OpenCV
        img = cv2.imread(os.path.join(folder_path3, filename))
        img = cv2.resize(img, (64, 64))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img.flatten()
        test_live.append(img)
        count += 1
        if count == 150:
            break
 
count = 0   
for filename in os.listdir(folder_path4):
    # Check if the file is an image
    if filename.endswith(".jpg") or filename.endswith(".png"):
        # Read the image file using OpenCV
        img = cv2.imread(os.path.join(folder_path4, filename))
        img = cv2.resize(img, (64, 64))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = img.flatten()
        test_fake.append(img)
        count += 1
        if count == 150:
            break
        
X_test = np.concatenate((test_live, test_fake))
y_test = np.concatenate((np.ones(len(test_live)), np.zeros(len(test_fake))))

# Train the SVM classifier
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = clf.predict(X_test)

# Calculate the accuracy, F1-score, and recall of the classifier
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)

# Print the results
print("Accuracy:", acc)
print("F1-score:", f1)
print("Recall:", rec)

# code for a Bagged Version of SVM
from sklearn.ensemble import BaggingClassifier
# Define the base SVM classifier
base_clf = svm.SVC(kernel='linear')

# Define the Bagging classifier with 10 SVM models
bag_clf = BaggingClassifier(estimator=base_clf, n_estimators=10, random_state=42)

# Fit the Bagging classifier on the training data
bag_clf.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = bag_clf.predict(X_test)

# Calculate the accuracy, F1-score, and recall of the classifier
acc_bagged = accuracy_score(y_test, y_pred)
f1_bagged = f1_score(y_test, y_pred)
rec_bagged = recall_score(y_test, y_pred)

# comparing the results
print(f"Accuracy of SVM is: {acc} vs the accuracy of bagged SVM is: {acc_bagged} ")
print(f"F1-score of SVM is: {f1} vs the F1-score of bagged SVM is: {f1_bagged} ")
print(f"Recall of SVM is: {rec} vs the Recall of bagged SVM is: {rec_bagged} ")

```
